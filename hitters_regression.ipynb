```python
df_act_pred = pd.read_csv("data_hitters_linear_weights.csv")
```

```python
df_act_pred.info()
```

```python
def predicted_run(row):
    return -411.81 + (0.451 * row['1B']) + (0.805 * row['2B']) + (1.072 * row['3B']) + (1.428 * row['HR']) + (0.326 * row['BB+HBP']) + (0.250 * row['SB']) # - (0.254 * [row['CS']])
```

```python
df_act_pred['prediction_2'] = df_act_pred.apply(predicted_run, axis=1)
```

```python
df_act_pred['absolute_error_2'] = df_act_pred.apply(lambda row: row['R']-row['prediction_2'], axis=1)
```

```python
df_act_pred
```

```python
np.mean(abs(df_act_pred['absolute_error']))
```
